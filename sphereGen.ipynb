```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# https://mathworld.wolfram.com/SpherePointPicking.html

def point_generator(n, r=1):
    # Obtain sphere coordinates
    u = np.random.uniform(0,1,n)
    v = np.random.uniform(0,1,n)
    theta = 2*np.pi*u
    phi = np.arccos(2*v-1)
    
    # Transforming to cartesian coordinates
    z = r*np.cos(phi)
    x = r*np.sin(phi)*np.cos(theta)
    #x = np.sqrt(1-np.power(np.cos(phi), 2)) * np.cos(theta)
    y = r*np.sin(phi)*np.sin(theta)
    #np.sqrt(1-np.power(np.cos(phi), 2)) * np.sin(theta)
    
    return x, y, z
```

```python
def visualize(x,y,z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(x,y,z, c='b', marker='o', s=20)
    
    ax.set_box_aspect([1,1,1])
    
    # Set labels and title
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    plt.title('Uniform Random Points on a Sphere')
    ax.view_init(azim=45, elev=30)
    
    plt.show()
```

```python
N = 1000
R = 1.0

x,y,z = point_generator(N, R)
visualize(x,y,z)
```

```python
import pickle

# Generate 10k points for a dataset
dataset = dict()
for i in range(10000):
    x, y, z = point_generator(N, R)
    dataset[i] = (x,y,z)

with open('sphereDataset.pickle', 'wb') as handle:
    pickle.dump(dataset, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
